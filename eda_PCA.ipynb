```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D

# 데이터 로드
feature_df = pd.read_csv('./feature.csv')
target_df = pd.read_csv('./target.csv')

# 결측치 처리
feature_df.fillna(feature_df.mean(), inplace=True)

# 상관 행렬 계산
correlation_matrix = feature_df.corr()

# 히트맵 시각화
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Feature Correlation Matrix')
plt.show()

# 상관계수가 0.9 이상인 피처 쌍 식별
high_corr_pairs = [(i, j) for i in correlation_matrix.columns for j in correlation_matrix.columns if i != j and abs(correlation_matrix.loc[i, j]) > 0.9]

# 중복 제거를 위해 첫 번째 피처만 선택
features_to_remove = set([j for i, j in high_corr_pairs])

# 피처 제거
reduced_feature_df = feature_df.drop(columns=features_to_remove)

# 데이터 스케일링
scaler = StandardScaler()
scaled_features = scaler.fit_transform(reduced_feature_df)

# PCA 수행
pca = PCA()
pca_features = pca.fit_transform(scaled_features)

# 1. 설명된 분산 비율 누적합 시각화
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_.cumsum(), marker='o', linestyle='--')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('PCA Explained Variance')
plt.grid()
plt.show()

# n_components=0.95로 설정하여 PCA 수행 (95% 설명 분산 비율을 유지하는 주성분 개수 선택)
pca = PCA(n_components=0.95)
pca_features_95 = pca.fit_transform(scaled_features)

# PCA 결과를 데이터프레임으로 변환
pca_feature_df = pd.DataFrame(pca_features_95, columns=[f'PC{i+1}' for i in range(pca_features_95.shape[1])])

# 2. PCA 2D 시각화
plt.figure(figsize=(10, 6))
plt.scatter(pca_features_95[:, 0], pca_features_95[:, 1], alpha=0.5)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA 2D Visualization')
plt.grid()
plt.show()

# 3. PCA 3D 시각화
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pca_features_95[:, 0], pca_features_95[:, 1], pca_features_95[:, 2], alpha=0.5)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title('PCA 3D Visualization')
plt.show()

# 4. 주성분 별 중요 변수 시각화
components = pd.DataFrame(pca.components_, columns=reduced_feature_df.columns)

plt.figure(figsize=(15, 15))
sns.heatmap(components, cmap='coolwarm', yticklabels=[f'PC{i+1}' for i in range(components.shape[0])])
plt.xlabel('Features')
plt.ylabel('Principal Components')
plt.title('PCA Components')
plt.show()
```
